```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from datasets import load_dataset

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')

print("Imports successful")
```

```python
dataset = load_dataset("stanfordnlp/imdb")
train_df = pd.DataFrame(dataset['train'])
test_df = pd.DataFrame(dataset['test'])
print("Data loaded")
```

```python
stop_words = set(stopwords.words('english'))

def clean_text(text):
    # 1. Lowercase
    text = text.lower()
    # 2. Remove HTML tags (IMDb reviews have these)
    text = re.sub(r'<.*?>', '', text)
    # 3. Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    # 4. Remove punctuation and special characters
    text = re.sub(r'[^a-z\s]', '', text)
    # 5. Tokenize
    tokens = word_tokenize(text)
    # 6. Remove stopwords
    tokens = [t for t in tokens if t not in stop_words]
    # 7. Rejoin
    return ' '.join(tokens)

print("Function defined")
```

```python
sample = train_df['text'][0]
print("ORIGINAL:\n", sample[:300])
print("\nCLEANED:\n", clean_text(sample)[:300])
```

```python
print("Cleaning training data... (this takes ~1-2 mins)")
train_df['clean_text'] = train_df['text'].apply(clean_text)

print("Cleaning test data...")
test_df['clean_text'] = test_df['text'].apply(clean_text)

print("Done!")
train_df[['text', 'clean_text', 'label']].head(3)
```

```python
train_df.to_csv('../data/train_clean.csv', index=False)
test_df.to_csv('../data/test_clean.csv', index=False)

print("Saved to data/ folder")
```
